```python
def draw_plot_median(attr):
    a = df[attr].unique()
    a.sort()
    plt.scatter(a, df.groupby(attr)['G3'].median())
    plt.show()
```

```python
def draw_plot_mean(attr):
    a = df[attr].unique()
    a.sort()
    plt.scatter(a, df.groupby(attr)['G3'].mean())
    plt.show()

```

```python
def draw_plot(attr):
    plt.scatter(df[attr], df['G3'])
    plt.show()
```

```python
def draw_plots(attr):
    draw_plot_mean(attr)
    draw_plot_median(attr)
    draw_plot(attr)
```
