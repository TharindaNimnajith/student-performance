```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/processed/student-performance-dataset.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
df['sex'].value_counts()
```

```python
fig = plt.figure()
ax = df['sex'].value_counts().plot(kind='pie',
                                   labels=['female', 'male'],
                                   autopct='%1.0f%%')
ax.yaxis.set_visible(False)
plt.title('Gender')
plt.legend()
fig.savefig('plots/gender.png', facecolor='white')
plt.show()
```

```python
df['age'].value_counts()
```

```python
fig = plt.figure()
df['age'].value_counts().plot(kind='bar')
plt.title('Age')
plt.xlabel('age')
plt.ylabel('count')
fig.savefig('plots/age.png', facecolor='white')
plt.show()
```
